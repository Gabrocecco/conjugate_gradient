```python
import pandas as pd
import matplotlib.pyplot as plt

# Matplotlib inâ€‘notebook
%matplotlib inline
```

```python
# === Random sparse matrices (mv_prof_random.csv) ===
old_csv = "data/mv_prof_random.csv"   

df_old = pd.read_csv(old_csv)

plt.figure(figsize=(8,5))
for sparsity, grp in df_old.groupby("sparsity"):
    g = grp.sort_values("n")
    plt.plot(g["n"], g["speedup_time"], marker='o', label=f"Sparsity {sparsity:.2f}")

plt.axhline(y=2.0, color='red', linestyle='--', linewidth=2)
plt.text(df_old["n"].max(), 2.03, "Theoretical max (2x)", fontsize=9, color='red', ha='right')

plt.axhline(y=1.0, color='grey', linestyle='--', linewidth=2)

plt.xlabel("Matrix size (n)")
plt.ylabel("Speedâ€‘up (scalar vs vectorized)")
plt.title("Random sparse matrices")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === OpenFOAM cylinder cases (mv_prof_foam.csv) ===
foam_csv = "data/mv_prof_foam.csv"   

df_foam = pd.read_csv(foam_csv)

plt.figure(figsize=(8,5))
g = df_foam.sort_values("n")
plt.plot(g["n"], g["speedup_time"], marker='o')

plt.axhline(y=2.0, color='red', linestyle='--', linewidth=2)
plt.text(g["n"].max(), 2.03, "Theoretical max (2x)", fontsize=9, color='red', ha='right')

plt.axhline(y=1.0, color='grey', linestyle='--', linewidth=2)
# plt.text(df_old["n"].max(), 1.03, "Scalar baseline", fontsize=9, color='blue', ha='right')

plt.xlabel("Matrix size (n)")
plt.ylabel("Speedâ€‘up (scalar vs vectorized)")
plt.title("OpenFOAM cylinder cases")
plt.grid(True)
plt.tight_layout()
plt.show()
```
